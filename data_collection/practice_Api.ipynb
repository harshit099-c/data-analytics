```python
def func(from_curr,to):
 try:
  url=f'https://api.frankfurter.dev/v1/latest?base={from_curr}&symbols={to}'
  response=requests.get(url)
  amount=int(input("Enter Amount"))
  data=response.json()
  converted_amt=amount*data['rates'][to]
  
  if 'rates' not in data or to not in data['rates']:
   print('Invalid Currency')
   return
  print(f"{converted_amt}{to}")

 except requests.exceptions.RequestException:
  print("Error in fetching data from api")

 except ValueError:
  print("Enter Valid Value")

func(input("Enter From"),input("Enter To"))
```
