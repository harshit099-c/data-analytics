```python
import json
def open_func(file_name):
    with open(file_name,"r") as f:
      
      data=json.load(f)
      return data
```

```python
data=open_func("print.json")
print(data)
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_data=set()

    for i in data:
        raw_rating=i.get("rating")
        if isinstance(raw_rating,str):
         raw_rating=i["rating"].strip().lower() #strip removes spaces
         if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]  
            i["rating"]=raw_rating     
        raw_age=i.get("age")
        if raw_age==None:
            i["age"]="None"
        
        if(i["name"].strip() in unique_data):
           continue

        unique_data.add(i["name"].strip())
        cleaned_data.append(i)

    with open("print.json","w") as f:
       json.dump(cleaned_data,f,indent=4)   

       

          
```

```python
clean_data(data)
```

```python
def rec(new_data):
    recommendation=[]

    
    for i in new_data:
        curr_rec={}
        curr_rec["name"]=i["name"]

        if(float(i["rating"])>4):
            curr_rec["brand"]="Apple"
        else:
            curr_rec["brand"]="Samsung"

        recommendation.append(curr_rec)

    return recommendation    
```
