```python
oscar_win=[
    "The Dark Knight",
    "Iron-Man",
    "Avengers-Endgame",
    "Stranger-Things",
    "Fast&Furious"
]

revenue=[1000,800,924,689,721]
year=[2011,2012,2019,2025,2017]

plt.plot(year,revenue)

plt.title("Oscar Movues Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()
```

```python
non_oscar=["Slumdog Millionaire","Avatar","Ant-Man","SpiderMan-farfromhome","Dr.Strange"]
revenue_nonoscar=[203,2034,590,900,320]
plt.plot(year,revenue,label="Oscar-Winning")
plt.plot(year,revenue_nonoscar,label="Non-Oscar")

plt.title("Comparison oscar vs non-oscar")
plt.xlabel("Years")
plt.ylabel("Revenue in($M)")
plt.legend()

plt.show()
```

```python
#Format strings [marker][line][color]
plt.plot(year,revenue,"o--r",label="Oscar-Winning",marker="<",color="#0af9f9",linestyle="-.",linewidth=3)
plt.plot(year,revenue_nonoscar,"<-y",label="Non-Oscar")

plt.title("Comparison oscar vs non-oscar")
plt.xlabel("Years")
plt.ylabel("Revenue in($M)")
plt.legend()
plt.show()
```

```python
#methods and attributes of format strings
#marker: o, <, >, ^, s, D, d, p, *, +, x
#line: --, -., :, -
#color: r, g, b, y, c, m, k, w
with plt.xkcd():

 plt.plot(year,revenue,label="Oscar-Winning")
 plt.plot(year,revenue_nonoscar,label="Non-Oscar")
 plt.title("Comparison oscar vs non-oscar")
 plt.xlabel("Years")
 plt.ylabel("Revenue in($M)")
 plt.legend()

 plt.style.available
 plt.style.use("ggplot")
 plt.grid()

 
 plt.show()
```

```python
with plt.xkcd():

    plt.plot(oscar_win,revenue)
    plt.title("Movie Revenue")
    plt.xlabel("Movies")
    plt.ylabel("Revenue (in $M)")
    plt.grid()
    plt.tight_layout()
    plt.style.use("ggplot")

    plt.show()
    plt.savefig("movies-revenue.png")
```

```python
plt.bar(year,revenue,width=1)
plt.title("Oscar movie revenue")
plt.xlabel("years")
plt.ylabel('Revenue (in $M)')

for i in range(len(year)):
 #bar labels
 plt.text(year[i],revenue[i],oscar_win[i],ha="center",va="bottom")

plt.ylim(0,max(revenue)+100)
plt.xlim(min(year)-1,max(year)+5)
plt.show()
```
