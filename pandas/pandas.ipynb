```python
ds=pd.Series([1,2,3,4,5])
ds
print(type(ds),ds) #dtype is also printed together
```

```python
#custom labels
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s
print(s['c'])
print(s.index)
```

```python
#vectorization=no loops drop(index)

s1=pd.Series([10,20,30,40])
s2=pd.Series([2,3,4,5])
print(s1+s2)

s1[0]=100
new_s1=s1.drop(0) #size immutable in pandas so new s1 is created
print(s1)
s2=s2.add(100)
print(s2)
print(new_s1)
```
