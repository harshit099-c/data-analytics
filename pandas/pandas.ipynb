```python
#transforming data

gender_map={"Male":"M","Female":"F","Unknown":"U"}
df_raw=pd.read_csv("raw_data.csv")
print(df_raw)    
df_raw['gender']=df_raw['gender'].map(gender_map).ffill()
print(df_raw)

df_raw['tax']=df_raw['income'].apply(lambda x:"20%" if x>=50000 else "10%")
df_raw['tax']

df_raw=df_raw.assign(new_income=df_raw['income']*1.1)
print(df_raw)

df_raw['country'].replace("USA",'US')
df_raw['income']=df_raw['new_income']
df_raw=df_raw.drop(columns=['new_income'])
print(df_raw['income'].ffill())
```

```python
df_raw.drop(columns=["tax"])
df_raw['tax']=df_raw['income'].apply(lambda x:20 if x>=60000 else 10)
df_raw

df_raw=df_raw.assign(income_after=df_raw['income']-(df_raw['income']*df_raw['tax']/100))
df_raw.drop(columns=['income'])


df_age=df_raw['age'].mean()
df_age

df_raw['age']=df_raw['age'].fillna(df_age)
df_raw['age']=df_raw['age'].astype(int)
df_raw.drop_duplicates(inplace=True)
df_raw
```

```python
df_raw2=df_raw.copy()

new_col=[col for col in df_raw.columns if col !='id']+['id']

df_raw2=df_raw2[new_col]
df_raw2
```

```python
df_copy=pd.read_csv("raw_data.csv")
df_copy

df_copy['tax(in %)']=df_copy['age'].apply(lambda x:20 if x>35 else 10)
df_copy.drop(columns=['tax(in %)'])

df_copy['tax(in %)']=df_copy['income'].apply(lambda x:20 if x>20 else 10)

#map
country_map={'USA':"US","Canada":"CAN","Mexico":"Mxco","India":"IN","Spain":"SPN"}
df_copy['country']=df_copy['country'].map(country_map)

#assign

df_copy=df_copy.assign(Income_in_hand=df_copy['income']-(df_copy['income']*df_copy['tax(in %)']/100))

#replace
df_copy=df_copy.replace("Male","M")
df_copy=df_copy.replace("Female","F")
df_copy=df_copy.replace("Unknown","U")

#rename
df_copy.columns=df_copy.columns.str.capitalize()
df_copy

df_copy.rename(columns={"Income":"Salary","Income_in_hand":"Salary_in_hand"},inplace=True)

#sort_values

df_copy.sort_values(['Salary_in_hand','Age'],ascending=False,inplace=True)

df_copy.drop_duplicates(inplace=True)

#sort_index

df_copy.sort_index()

#reset index

df_copy.reset_index(drop=True,inplace=True)
df_copy
#rank

df_copy['Rank']=df_copy['Salary'].rank(method="dense",ascending=False)

df_copy[['Salary','Salary_in_hand','Rank']]=df_copy[['Salary','Salary_in_hand','Rank']].ffill()
df_copy

df_copy[['Age']]=df_copy[['Age']].ffill()

df_copy[['Age','Salary','Salary_in_hand','Rank']]=df_copy[['Age','Salary','Salary_in_hand','Rank']].astype(int)
df_copy

#column reordering

new_columns= [col for col in df_copy.columns if col!='Id']+['Id'] #concate

df_copy=df_copy[new_columns]
print(df_copy)
```

```python
#writing in csv
df_copy.to_csv("raw_data.csv") #modify existing file
```
