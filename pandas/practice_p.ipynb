```python
import pandas as pd
```

```python
df_petal=pd.read_csv('IRIS.csv')

#1
df_petal.head(10)
print(df_petal.shape,df_petal.dtypes)

df_petal.describe()

#2

df_petal[(df_petal['petal_length']>4.5) & (df_petal['species']=='Iris-virginica')]

#3
df_petal.groupby('species').agg(
    avg_length=('sepal_length','mean'),
    max_width=('petal_width','max'),
    standard_dev=('sepal_width','std')
)

df_petal=df_petal.assign(petal_ratio=df_petal['petal_length']/df_petal['petal_width'])

df_petal.groupby('species')['petal_ratio'].mean()
```

```python
#1
df_tit=pd.read_csv('Titanic-Dataset.csv')
df_tit.columns

#2
df_tit[(df_tit['Sex']=='female') & (df_tit['Fare']>30)]

#3
df_tit.groupby('Pclass').agg(
    Survial_rate=('Survived',lambda x:x.mean()*100),
    Average_Fare=('Fare','mean'),
    Average_Age=('Age','mean')
)
```
