```python
#Melt & Pivot

import pandas as pd

df_copy=pd.read_csv('raw_data.csv')
df_copy['Name']=df_copy['Name'].bfill(inplace=True)
df_copy=df_copy.reset_index(drop=True)
df_copy.to_csv('raw_data.csv')
```

```python
#melt and pivot
melted_df=df_copy.melt(        #expand data in rows
    id_vars=['Name','Id'],
    value_vars=['Salary','Salary_in_hand'],
    var_name='metrics',
    value_name='value'
)

melted_df.pivot( #expand in columns or back from melt
    index=['Name','Id'],
    columns='metrics',
    values='value'
)
```

```python
#Merging and Joining
df_customers=pd.DataFrame({
    'cust_id':[1,2,3,4],
    'cust_name':['Harshit','Hua','Ksj','Gss']

})

df_orders=pd.DataFrame({
    'order_id':[101,102,103,104],
    'amount':[200,300,399,400],
    'cust_id':[2,1,4,5]
})
```

```python
#inner join

pd.merge(df_customers,df_orders,on='cust_id')
pd.merge(df_customers,df_orders,on='cust_id',how='left')
pd.merge(df_customers,df_orders,on='cust_id',how='right')
pd.merge(df_customers,df_orders,on='cust_id',how='outer') #union
```

```python
#data concatenation

pd.concat([df_customers,df_orders]) #rowise concat

#to reset row labels
pd.concat([df_customers,df_orders],ignore_index=True)

#columnwise concat
pd.concat([df_customers,df_orders],axis=1)
```
